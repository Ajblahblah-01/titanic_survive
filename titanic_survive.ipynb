```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.linear_model import LogisticRegression 
from sklearn.metrics import classification_report, confusion_matrix
```

```python
df = pd.read_csv("/content/drive/MyDrive/titanic_survive/train.csv")
df.head()
```

```python
df = df.drop(['Cabin','Name','Ticket'],axis = 1)
df.head()
```

```python
def gender_to_binary(sex):
    if sex == 'male':
        return 0
    else:
        return 1;

df['Sex'] = df['Sex'].apply(gender_to_binary)
df.head()
```

```python
def embark_to_num(emb):
    if emb == 'S':
        return 0
    elif emb == 'C':
        return 1
    else:
        return 2

df['Embarked'] = df['Embarked'].apply(embark_to_num)
df.head()
```

```python
df = df.drop(['PassengerId'],axis = 1)
df.head()
```

```python
df['Age'].fillna(df['Age'].mean(),inplace = True)
```

```python
# input
x = df.iloc[:, [1,2,3,4,5,6,7]].values

# output
y = df.iloc[:, 0].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
xtrain = sc_x.fit_transform(X_train)
xtest = sc_x.transform(X_test)

print (xtrain[0:10, :])
```

```python
classifier = LogisticRegression(random_state = 0,solver='lbfgs',max_iter = 1000)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
cm = confusion_matrix(y_test, y_pred)
 
print ("Confusion Matrix : \n", cm)
```

```python

from sklearn.metrics import accuracy_score
print ("Accuracy : ", accuracy_score(y_test, y_pred))
```
